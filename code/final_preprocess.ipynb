```python
import PyPDF2
import numpy as np
import pandas as pd

# Import necessary modules
import ssl
from nltk.tokenize import sent_tokenize
from nltk.tokenize import word_tokenize
ssl._create_default_https_context = ssl._create_unverified_context

import nltk
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

# Import WordNetLemmatizer, and stopwords
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords

# import spcay framework
import spacy
nlp = spacy.load('en_core_web_sm')
from spacy.symbols import amod
```

```python
path = "/Users/hs/Documents/SDAIA_DS_T5/Project04-NLP/experiments/data/all_chapters.pdf"
ff = open( path ,"rb")
pdfReader = PyPDF2.PdfFileReader(ff)
print(pdfReader.numPages)
total_page = pdfReader.numPages
```

```python
# convert the  article value from list type to str type in order to tokenize it.

# Function to convert  
def listToString(s): 
    
    # initialize an empty string
    str1 = " " 
    
    # return string  
    return (str1.join(s))
```

```python
article = " "

for page_num  in np.arange(0 , total_page):
    pageObj = pdfReader.getPage(page_num)
    extract = pageObj.extractText().split("\n")
    article += listToString(extract)
```

```python
type(article)
```

```python
# Import Counter
from collections import Counter

# Tokenize the article: tokens
tokens = word_tokenize(article)

# Convert the tokens into lowercase: lower_tokens
lower_tokens = [t.lower() for t in tokens]

# Create a Counter with the lowercase tokens: bow_simple
bow_simple = Counter(lower_tokens)

# Print the 10 most common tokens
print(bow_simple.most_common(10))
```

```python
# Retain alphabetic words: alpha_only
alpha_only = [t for t in lower_tokens if t.isalpha()]

# Remove all stop words: no_stops
no_stops = [t for t in alpha_only if t not in stopwords.words('english')]

# Instantiate the WordNetLemmatizer
wordnet_lemmatizer = WordNetLemmatizer()

# Lemmatize all tokens into a new list: lemmatized
lemmatized = [wordnet_lemmatizer.lemmatize(t) for t in no_stops]

# Create the bag-of-words: bow
bow = Counter(lemmatized)

# Print the 10 most common tokens
print(bow.most_common(20))
```

```python
lsa = TruncatedSVD(3)
doc_topic = lsa.fit_transform(doc_word)
print(lsa.explained_variance_ratio_)
```

```python
topic_word = pd.DataFrame(lsa.components_.round(3),
             index = ['component'+str(i) for i in range(3)],
             columns = vectorizer.get_feature_names())

print(topic_word)
```

```python
def display_topics(model, feature_names, no_top_words, topic_names=None):
    for ix, topic in enumerate(model.components_):
        if not topic_names or not topic_names[ix]:
            print("\nTopic ", ix)
        else:
            print("\nTopic: '",topic_names[ix],"'")
        print(", ".join([feature_names[i]
                        for i in topic.argsort()[:-no_top_words - 1:-1]]))
```

```python
display_topics(lsa, vectorizer.get_feature_names(), 30)
```
